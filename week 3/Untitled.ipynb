```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

def verify_employee():
    name_input = name_entry.get().strip()
    dept_input = dept_entry.get().strip()
    
    try:
        # Load the dataset
        df = pd.read_csv('GIG-logistics.csv')
        
        # Check if employee exists in that department
        match = df[(df['Name'].str.lower() == name_input.lower()) & 
                   (df['Department'].str.lower() == dept_input.lower())]
        
        if not match.empty:
            # Get other members of the same department
            team = df[df['Department'].str.lower() == dept_input.lower()]['Name'].tolist()
            team.remove(match.iloc[0]['Name']) # Remove the current user from the list
            
            result_msg = f"Welcome {name_input}!\n\nYour team members in {dept_input}:\n" + "\n".join(team)
            messagebox.showinfo("Success", result_msg)
        else:
            messagebox.showerror("Error", "Employee does not exist in this department.")
            
    except FileNotFoundError:
        messagebox.showerror("Error", "Dataset file 'GIG-logistics.csv' not found.")

# GUI Setup
root = tk.Tk()
root.title("GIG Logistics Verification")
root.geometry("300x200")

tk.Label(root, text="Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department:").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Verify", command=verify_employee).pack(pady=20)

root.mainloop()
```
