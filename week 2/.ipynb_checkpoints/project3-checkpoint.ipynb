  1. Create lists for Girls: Names, Ages, Heights, Scores.
  2. Create lists for Boys: Names, Ages, Heights, Scores.
  3. Combine all Names, Ages, Heights, and Scores into a single dataset.
  4. Use Pandas to create a 'DataFrame' (Table) using these lists.
  5. Print the DataFrame in a tabular format.
STOP

```python
import pandas as pd

girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
girls_age = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
girls_height = [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
girls_scores = [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]

boys_names = ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"]
boys_age = [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
boys_height = [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
boys_scores = [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]

all_names = girls_names + boys_names
all_ages = girls_age + boys_age
all_heights = girls_height + boys_height
all_scores = girls_scores + boys_scores

data = {
    'Name': all_names,
    'Age': all_ages,
    'Height': all_heights,
    'Score': all_scores
}

df = pd.DataFrame(data)

print("--- Class Student Records ---")
print(df.to_string(index=False)) # index=False hides the row numbers
```
