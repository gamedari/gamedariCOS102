```python
import numpy as np

def solve_equation():
    print("Polynomial Equation Solver")
    print("1. Quadratic (Ax² + Bx + C = 0)")
    print("2. Cubic (Ax³ + Bx² + Cx + D = 0)")
    print("3. Quartic (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
    
    choice = input("\nSelect the operation (1, 2, or 3): ")

    if choice == '1':
        print("Enter coefficients for Ax² + Bx + C = 0")
        a = float(input("A: "))
        b = float(input("B: "))
        c = float(input("C: "))
        coeffs = [a, b, c]
    
    elif choice == '2':
        print("Enter coefficients for Ax³ + Bx² + Cx + D = 0")
        a = float(input("A: "))
        b = float(input("B: "))
        c = float(input("C: "))
        d = float(input("D: "))
        coeffs = [a, b, c, d]
        
    elif choice == '3':
        print("Enter coefficients for Ax⁴ + Bx³ + Cx² + Dx + E = 0")
        a = float(input("A: "))
        b = float(input("B: "))
        c = float(input("C: "))
        d = float(input("D: "))
        e = float(input("E: "))
        coeffs = [a, b, c, d, e]
    
    else:
        print("Invalid selection.")
        return

    roots = np.roots(coeffs)
    
    print("\nThe roots are:")
    for i in roots:
        print(np.round(i, 4))

solve_equation()
```
