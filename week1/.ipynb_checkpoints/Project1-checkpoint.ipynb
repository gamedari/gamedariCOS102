```python
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
T = float(input("Enter Time (T): "))

A = P * (1 + (R / 100) * T)
SI = A - P

print(f"Amount: {A}")
print(f"Simple Interest: {SI}")

![My Image](SI.png)
```

```python
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
N = float(input("Enter Number of Compounding Periods per Year (N): "))
T = float(input("Enter Time (T): "))

A = P * (1 + R / N) ** (N * T)
CI = A - P

print(f"Amount: {A}")
print(f"Compound Interest: {CI}")


![My Image](CI.png)
```

```python
PMT = float(input("Enter Periodic Payment Amount (PMT): "))
R = float(input("Enter Rate (R): "))
N = float(input("Enter Number of Compounding Periods per Year (N): "))
T = float(input("Enter Time (T): "))

A = PMT * (((1 + R / N) ** (N * T) - 1) / (R / N))

print(f"Annuity Amount (A): {A}")

![My Image](AN.png)
```
